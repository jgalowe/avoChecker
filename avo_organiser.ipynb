### Workflow to classify new avocadoes
Step 1: Rename the pictures in the folder

```python
import os
from ultralytics import YOLO
model_yolo = YOLO("avoID.pt")
```

```python
#establishing the folder name which we will use later in our renaming process
directory = 'Pictures/1030B'
folder_name = os.path.basename(directory)

index = 0
for item in os.listdir(directory):
    os.rename((directory + '/' + item), 
              (directory + '/' + folder_name + f'_{index}.jpg')), 
              
    index=index+1
```

Step 2: generate the crops for the avocados

```python
model_yolo.predict(
        source=directory,
        save_crop=True,
        save=True,
        conf=0.5
    )
```
